```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate


def Simpson(begin, end, partitions):
    '''
    Approximates the area under the curve using series expansion
    '''
    f = lambda x: (x**4 * np.exp(x)) * (np.exp(x)-1)**-2
    
    dx = (end - begin)/partitions
    X = np.linspace(begin, end, partitions+1)
    Y = f(X)
    
    area = np.zeros([partitions+1],float)
    for i in range (partitions+1):
        if i==0 or i==partitions:
            coef = 1/3
        elif i%2 == 1:
            coef = 2/3
        else:
            coef = 4/3
        area[i] = coef * Y[i] * dx
        
    return sum(area)


def heat_capacity(start, end, subintervals,plot=1):
    '''
    Program to calculate the heat capacity of a solid
    '''
    
    #constants
    V = 0.001              #Volume in meters cubed
    rho = 6.022 * 10**28   #Density of aluminum in kg/m^3
    kb = 1.3807 * 10**-23  #Boltzman constant J/K
    theta = 428            #Debye temperature in K
    
    #array for x values (temperature)
    T = np.linspace(start, end, subintervals)
    
    #array for y values (Cv)
    Cv = np.empty([subintervals], float)
    
    #calculate the heat capacity (Cv)
    for i in range(subintervals):
        T[i] = 5 * i                 #for temperature range
        Cv[i] = 9*V*rho*kb*(T[i]/theta)**3 * Simpson(start,end,subintervals+1) #for cv calculation
    
    #now make plot
    plt.plot(T, Cv, 99)    
    plt.xlim([start,end])
    plt.ylim([min(Cv),max(Cv)])
    plt.title('Heat Capacity of a Solid from 5K - 500K')
    plt.xlabel('Temperature (K)')
    plt.ylabel('Heat Capacity (Cv)')
    plt.show()
```

```python
heat_capacity(5,500,99)
```
