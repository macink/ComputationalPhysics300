Now we can apply the "One Rule" algorith based on frequency.
This is a classification algorithm based on one rule.
We generate one rule for each predictor in our data and then choose the rule that has the smallest total error and this is our one rule.

For each predictor:

Step 1: for each value of that predictor make a rule:

    -Count how often each value of target (class) appears
    -Find the most frequent class
    -Make the rule assign that class to this value of the predictor
    
Step 2: Calculate the total error of the rules of each predictor

Step 3: Choose the predictor with the smallest total error
