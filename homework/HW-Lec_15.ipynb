```python
%matplotlib notebook
#code to solve minimum of 2D function with Monte Carlo method

#First let's check out the plot
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#set up range
x_min, x_max = -2,2
y_min, y_max = -2,2

#function to plot and minimize
def f(x,y):
    return x**2/2 + y**2/3 - x*y/4

#arrays to hold grid coordinates
nx = np.linspace(x_min, x_max, 100)
ny = np.linspace(y_min, y_max, 100)

#array for values from function
x,y = np.meshgrid(nx,ny)

z = f(x,y)

#set up plot
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(x,y,z, cmap='summer', alpha=0.8)
cset = ax.contour(x,y,z, zdir='z', offset=0)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('$f(x,y)$')

plt.show
```
