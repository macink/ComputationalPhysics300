```python
#Program that uses Faster Fourier Transformation to process image 
#and compress the data

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.figure as fig

data = np.loadtxt("circular.txt")
print(np.shape(data))
plt.imshow(data)
plt.show()
plt.savefig("Lec_09_not_compressed")
fig.figaspect(data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.figure as fig

data = np.loadtxt("circular.txt")

plt.plot(data)
plt.show()

x = np.fft.fft(data)
plt.plot(np.abs(x))
plt.show()
plt.savefig("fft")
fig.figaspect(x)
```
